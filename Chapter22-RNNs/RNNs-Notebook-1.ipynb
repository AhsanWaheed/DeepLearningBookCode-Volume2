### Notebook 1: Repeated activation functions

This notebook is provided as a “behind-the-scenes” look at code used to make some of the figures in this chapter. It is still in the hacked-together form used to develop the figures, and is only lightly commented.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make a File_Helper for saving and loading files.

save_files = True

import os, sys, inspect
current_dir = os.path.dirname(os.path.abspath(inspect.getfile(inspect.currentframe())))
sys.path.insert(0, os.path.dirname(current_dir)) # path to parent dir
from DLBasics_Utilities import File_Helper
file_helper = File_Helper(save_files)
```

```python
def repeat_tanh(x, n):
    v = x
    for i in range(0, n):
        v = np.tanh(v)
    return v
```

```python
xs = np.linspace(-8, 8, 500)
plt.plot(xs, repeat_tanh(xs, 1), lw=2, c='black', linestyle='-', label="1 tanh")
plt.plot(xs, repeat_tanh(xs, 2), lw=2, c='black', linestyle='--', label="2 tanh")
plt.plot(xs, repeat_tanh(xs, 5), lw=2, c='black', linestyle='-.', label="5 tanh")
plt.plot(xs, repeat_tanh(xs, 25), lw=2, c='black', linestyle=':', label="25 tanh")
plt.legend(loc='best')
plt.xlabel("input")
plt.ylabel("output")
plt.title("Repeated tanh")
file_helper.save_figure('repeated-tanh')
plt.show()
```
